```python
import numpy as np

def f(x):
    return 2 * x + 1

# x ni 3 ga ikki tomondan yaqinlashtiramiz
for x in [2.9, 2.99, 2.999, 3.001]:
    print(x, '->', f(x))

# 2.9   -> 6.8
# 2.999 -> 6.998
# 3.001 -> 7.002    <- 7 ga intilmoqda!
```

```python
def f(x):
    return (x**2 - 4) / (x - 2)

# f(2) -> ZeroDivisionError!

for x in [1.9, 1.99, 1.999, 2.001]:
    print(x, f(x))

# 1.9   3.9
# 1.999 3.999
# 2.001 4.001   <- limit = 4!
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.abs(x) / x

# chapdan va o'ngdan jadval
for x in [-0.01, -0.001, 0.001, 0.01]:
    print(x, f(x)) # -1, -1, +1, +1

# grafik: sakrashni ko'ramiz
xs = np.linspace(-2, 2, 55)
xs = xs[xs != 0] # 0 ni chiqarib tashla (chunki 0 ga bo'lib bo'lmaydi)

plt.scatter(xs, f(xs), alpha=0.3, s=50) # rasmda xs/f(xs) o'rniga f(xs) berilgan
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# g(x) ya'ni ReLU funksiyasi
def g(x):
    return (x + np.abs(x)) / 2

# X qiymatlari to'plami
xs = np.linspace(-3, 3, 60)

# Scatter orqali grafik chizish
plt.scatter(xs, g(xs), color='blue', alpha=0.7, s=30)
plt.title("ReLU activation function: g(x) = (x + |x|) / 2")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid(True)
plt.show()
```
