```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
# Topshiriq 1 — Titanic: Yosh va Nar
df_titanic = pd.read_csv('/content/train-2.csv').dropna(subset=['Age', 'Fare'])
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_titanic, x='Age', y='Fare', hue='Pclass', palette='Set1')
plt.title("Yosh va chipta narxi orasidagi munosabat")
plt.xlabel("Yosh (Age)")
plt.ylabel("Chipta narxi (Fare)")
plt.show()
r_titanic = np.corrcoef(df_titanic['Age'], df_titanic['Fare'])[0, 1]
print(f'Titanic: r = {r_titanic:.4f}')
# Topshiriq 2 — Diamonds: Og'irlik va Nar
df_diamonds = sns.load_dataset('diamonds')

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_diamonds, x='carat', y='price', alpha=0.1)
plt.title("Carat vs Price (To'liq dataset)")
plt.show()

df_sample = df_diamonds.sample(1000, random_state=42)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_sample, x='carat', y='price', hue='cut')
plt.title("Carat vs Price (1000 ta namuna, Cut bo'yicha)")
plt.show()

r_original = np.corrcoef(df_diamonds['carat'], df_diamonds['price'])[0, 1]
r_log = np.corrcoef(np.log1p(df_diamonds['carat']), np.log1p(df_diamonds['price']))[0, 1]

print(f'Diamonds Original Pearson r = {r_original:.4f}')
print(f'Diamonds Log transform bilan r = {r_log:.4f}')
# Topshiriq 3 — Diamonds: Korrelyatsiya Heatma
numeric_cols = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
corr_matrix = df_diamonds[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='RdYlGn', vmin=-1, vmax=1)
plt.title("Diamonds Numeric Ustunlar Heatmap'i")
plt.show()

corr_with_price = corr_matrix['price'].sort_values(ascending=False)
print("Price bilan korrelyatsiya:")
print(corr_with_price)
# Topshiriq 4 — Tips: Regplot va Guruhla
df_tips = sns.load_dataset('tips')

plt.figure(figsize=(8, 5))
sns.regplot(data=df_tips, x='total_bill', y='tip')
plt.title("Umumiy hisob va Choycha (Total Bill vs Tip)")
plt.show()

sns.lmplot(data=df_tips, x='total_bill', y='tip', hue='time', height=6, aspect=1.3)
plt.title("Lunch va Dinner uchun Alohida Trend")
plt.show()

lunch  = df_tips[df_tips['time'] == 'Lunch']
dinner = df_tips[df_tips['time'] == 'Dinner']

r_lunch,  p_lunch  = stats.pearsonr(lunch['total_bill'],  lunch['tip'])
r_dinner, p_dinner = stats.pearsonr(dinner['total_bill'], dinner['tip'])

print(f'Lunch:  r={r_lunch:.4f}, p={p_lunch:.4f}')
print(f'Dinner: r={r_dinner:.4f}, p={p_dinner:.4f}')
# Topshiriq 5 — Flights: Pairplot va Tren
df_flights = sns.load_dataset('flights')

plt.figure(figsize=(8, 5))
sns.regplot(data=df_flights, x='year', y='passengers')
plt.title("Yillar bo'yicha Yo'lovchilar Trendi")
plt.show()

monthly_avg = df_flights.groupby('month', observed=False)['passengers'].mean().reset_index()
plt.figure(figsize=(10, 5))
sns.barplot(data=monthly_avg, x='month', y='passengers', palette='Blues_r')
plt.title("Oylar bo'yicha O'rtacha Yo'lovchilar Soni")
plt.show()

p_r, _ = stats.pearsonr(df_flights['year'], df_flights['passengers'])
s_r, _ = stats.spearmanr(df_flights['year'], df_flights['passengers'])
print(f'Pearson r: {p_r:.4f}')
print(f'Spearman r: {s_r:.4f}')

max_idx = df_flights['passengers'].idxmax()
min_idx = df_flights['passengers'].idxmin()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_flights, x='year', y='passengers', color='blue', label='Boshqa oylar')
plt.scatter(df_flights.loc[max_idx, 'year'], df_flights.loc[max_idx, 'passengers'], color='red', s=150, zorder=5, label='Eng ko\'p')
plt.scatter(df_flights.loc[min_idx, 'year'], df_flights.loc[min_idx, 'passengers'], color='green', s=150, zorder=5, label='Eng kam')
plt.title("Yillar bo'yicha Yo'lovchilar (Ekstremum nuqtalar bilan)")
plt.legend()
plt.show()
```
