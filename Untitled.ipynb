````python
# 2-bosqichning Statistik tahlil (Hypothesis Testing) qismi
from scipy import stats
# Pitsa narxini normallikka tekshiramiz
stat, p = stats.shapiro(df['total_price'].sample(500, random_state=42))
print(f'p-value = {p:.4f}') # Bu yerda ham p = 0.0000 chiqadi, ya'ni "Normal emas"
````

````python
narx = df['total_price']
n = len(narx)
mean = narx.mean()
se = stats.sem(narx)
ci = stats.t.interval(0.95, df=n-1, loc=mean, scale=se)
print(f"95% Ishonch oralig'i (CI): {ci}")
# Chiqqan natija (masalan, 16.50, 17.10) kelajakdagi o'rtacha xarid shu oraliqda bo'lishini isbotlaydi.
````

````python
classic_prices = df[df['pizza_category'] == 'Classic']['total_price']
chicken_prices = df[df['pizza_category'] == 'Chicken']['total_price']

t_stat, p = stats.ttest_ind(classic_prices, chicken_prices)
# Agar p < 0.05 chiqsa: "Kategoriyalar o'rtasidagi narx farqi tasodifiy emas, Chicken rostdan ham qimmat!" bo'ladi.
````

bu 3 ta kichik qadamlarda (tats.shapiro,ci,ttest)da

> 1-yacheyka: Shapiro-Wilk testi (Normallikni tekshirish)
Nima qildik? Kompyuterga pitsa narxlaridan random qilib 500 tasini berdik va: "Shu narxlar dunyodagi ideal normal taqsimotga (ya'ni o'rtasi baland, ikki cheti bir xil silliq tushgan qo'ng'iroqsimon grafikga) o'xshaydimi?" deb savol berdik.
Natija nima chiqdi? p-value = 0.0000 (ya'ni 0%).
Ma'nosi nima? Bu degani: "Ushbu pitsa narxlarining ideal normal holatdan og'ib, qiyshiq bo'lib ketishi shunchaki tasodif emas! Buning ehtimoli 0%!".
Yakuniy xulosa: Narxlarimiz Normal EMAS. Bu bizga Linear Regression (Chiziqli model) bu loyihada yomon ishlashini, uning o'rniga Random Forest tanlaganimiz 100% to'g'ri ekanini matematik isbotlab berdi.


2-yacheyka: 95% Ishonch oralig'i (Confidence Interval)
Nima qildik? Qo'limizda 48,620 ta buyurtma bor. Biz matematik formuladan: "Menga shunday bir chegara topib berki, men 95% ishonch bilan kelajakda keladigan hamma mijozlarning o'rtacha xaridi ham aynan shu oraliqqa tushadi deb ayta olay" deb so'radik.
Natija nima chiqdi? (16.78, 16.86) dollar.
Ma'nosi nima? Tasavvur qiling, pitsa do'koniga kelgusi oylarda yana millionlab mijozlar keladi. Biz hozirgi ma'lumotga qarab, 95% ishonch bilan ayta olamizki, o'sha kelajakdagi xaridorlarning ham umumiy o'rtacha cheki 16.78$ dan kam bo'lmaydi va 16.86$ dan ko'p bo'lmaydi. Oraliq juda tor, demak biznesimiz barqaror va narxlar bir maromda ketyapti.


3-yacheyka: Mustaqil T-test (Ikki guruhni solishtirish)
Nima qildik? Datasetdan faqat Classic pitsalarining narxlarini va faqat Chicken (tovuqli) pitsalarining narxlarini ajratib oldik. Ularni bir-biri bilan urishtirib (solishtirib): "Bu ikki turning o'rtacha narxi o'rtasidagi farq rostdan ham jiddiy biznes-farqmi yoki shunchaki tasodifiy o'zgarishmi?" deb so'radik.
Kutilayotgan natija (kodni print qilganingizda): Bu yerda ham p-value 0.05 dan ancha kichik chiqadi (masalan, 0.0000).
Ma'nosi nima? Agar siz yozgan if p < alpha: sharti bajarilsa, bu kompyuter tilida: "Eshat, bu ikki turdagi pitsaning narxlari o'rtasidagi farq shunchaki tasodif emas! Chicken pitsalari menyuda rostdan ham Classic'dan ko'ra sezilarli darajada qimmatroq sotiladi" degan xulosani beradi.


Bu 3 ta qadam model qurishdan oldin ma'lumotlarning ichki qonuniyatlarini tushunish uchun qilindi. Biz narxlar normal emasligini, kelajakdagi o'rtacha narx qanday bo'lishini va pitsa turlari o'rtasida rostdan ham narx farqi borligini fanning tili (statistika) bilan tasdiqlab oldik.
