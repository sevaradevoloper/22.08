```python
# 2-topshiriq (10 ball) — Data Visualization
# Seaborn yoki Matplotlib yordamida `ball` ustuni taqsimotini ko'rsatuvchi histogram, hamda `yosh` va `ball`
# o'rtasidagi bog'liqlikni ko'rsatuvchi scatter plot chizadigan kodni yozing. Grafiklarga sarlavha va o'qlar nomini qo'shing.

plt.figure(figsize=(8,5))
sns.histplot(
    df['math_score'],
    kde=True,
    color='red',
)
plt.title('`ball` ustuni taqsimoti')
plt.xlabel('matematika baholari')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data= df,
   x = 'math_score',
   y= 'reading_score',
   alpha=0.5,
    hue='writing_score'

)
plt.title("`yosh` va `ball` o'rtasidagi bog'liqlik")
plt.xlabel('matematika baholari')
plt.ylabel('õqish baholari')
plt.show()
```

```python

data = [12, 15, 18, 22, 25, 30, 30, 35, 40, 100]
df = pd.DataFrame(data, columns=['Qiymatlar'])

mean_val = df['Qiymatlar'].mean()
median_val = df['Qiymatlar'].median()
var_val = df['Qiymatlar'].var()
std_val = df['Qiymatlar'].std()

print(f"ǒrtacha qiymat (Mean): {mean_val}")
print(f"Mediana: {median_val}")
print(f"Dispersiya (Variance): {var_val:.2f}")
print(f"Standart og'ish (Standard Deviation): {std_val:.2f}")
```
