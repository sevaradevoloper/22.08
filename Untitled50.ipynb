```python
import numpy as np
D = np.diag([2,3])
print(D)
v=np.array([2,1])
print(D@v)

# (2×2)+(0×2)=4+0=4
# (0×1)+(3×1)=0+3=3
```

```python
import numpy as np

# 3x3 o'lchamli birlik matritsa yaratish
birlik_matritsa = np.eye(3)

print("Birlik Matritsa:\n", birlik_matritsa)
```

```python
I = np.eye(3)

print(I)

A=np.array([[5,2,1,],
           [0,3,4],
           [1,1,2]])
print(np.allclose(A @ I,A))
# Matritsani Birlik matritsaga ko'paytirish — sonni 1 ga ko'paytirish bilan bir xil. Natija hech qachon o'zgarmaydi: A×I=A.
```

```python
import numpy as np

# 2x2 matritsa B
B = np.array([[5, 3],
              [1, 2]])

I = np.eye(2)

print("B @ I:\n", B @ I)
print("I @ B:\n", I @ B)
# Ikkalasi ham B matritsaning o'zini qaytaradi!

print(np.allclose(B @ I,B))
print(np.allclose(I @ B,B))
print((2*I)@B)
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

# Teskari matritsani hisoblash
A_teskari = np.linalg.inv(A)

print("Asl Matritsa:\n", A)
print("\nTeskari Matritsa:\n", A_teskari)

# Tekshirib ko'ramiz: A va A_teskari ko'paytmasi birlik matritsa berishi kerak
print("\nKo'paytma (Birlik matritsa chiqishi kerak):\n", np.dot(A, A_teskari))
```

```python
# A=np.array([[5,2,1,],
#            [0,3,4],
#            [1,1,2]])

# print(np.linalg.det(A))
# A_teskari = np.linalg.inv(A)
# print(A_teskari)
# print(np.round(A@ A_teskari))



# kv_matritsa = np.array([[1, 4],
#                         [3, 7]])


# B = np.array([[2, 1],
#               [4, 2]]) # 2-qator 1-qatorning 2 barobari


# print(np.linalg.det(kv_matritsa))
# kv_teskari = np.linalg.inv(kv_matritsa)
# print(kv_teskari)
# print(np.round(kv_matritsa @ kv_teskari))


# print(np.linalg.det(B))
# B_teskari = np.linalg.inv(B)
# print(B_teskari)
# print(np.round(B @ B_teskari))


import numpy as np

def teskarilanadimi(matrix):
    # O'lchamlarni tekshirish (1-band: Kvadrat matritsaligini aniqlash)
    # matrix.shape bizga (qatorlar, ustunlar) ni beradi
    if len(matrix.shape) != 2 or matrix.shape[0] != matrix.shape[1]:
        return "Xatolik: Matritsa kvadrat emas! Faqat kvadrat matritsalarni teskarilash mumkin."

    # 2. Determinantni hisoblash
    det = np.linalg.det(matrix)

    # Kompyuter float xatoliklari sababli 0 ni 0.0000001 deb hisoblab yubormasligi uchun
    # isclose funksiyasi orqali nolga yaqinligini tekshiramiz
    if np.isclose(det, 0):
        return f"Xatolik: det = 0. Bu singular (maxsus) matritsa, chunki ustunlari o'zaro bog'liq!"

    # 3. Agar hammasi joyida bo'lsa, teskari matritsani qaytaramiz
    teskari_m = np.linalg.inv(matrix)
    return teskari_m

# ---- UCH XIL MATRITSA BILAN SINAB KO'RAMIZ (3 va 4-bandlar) ----

# 1. Oddiy kvadrat matritsa
A = np.array([[3, 2],
              [1, 4]])

# 2. Singular (bog'liq ustunli) matritsa
B = np.array([[2, 6],
              [1, 3]])

# 3. Kvadrat bo'lmagan matritsa (2 qator, 3 ustun)
C = np.array([[1, 2, 3],
              [4, 5, 6]])

print("--- A matritsa testi ---")
res_A = teskarilanadimi(A)
print(res_A)
if isinstance(res_A, np.ndarray): # Agar natija matritsa bo'lsa, allclose bilan tekshiramiz
    print("Tekshirish (A @ A_inv = I):", np.allclose(A @ res_A, np.eye(A.shape[0])))

print("\n--- B matritsa testi ---")
print(teskarilanadimi(B))

print("\n--- C matritsa testi ---")
print(teskarilanadimi(C))
```
