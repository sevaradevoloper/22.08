```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

population = np.random.exponential(scale=2.0, size=10000)

sample_sizes = [5, 30, 100]
num_samples = 1000

pop_mean = np.mean(population)
pop_std = np.std(population)

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=False)

for i, n in enumerate(sample_sizes):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=n, replace=False)
        sample_means.append(np.mean(sample))

    sample_means = np.array(sample_means)

    axes[i].hist(sample_means, bins=30, edgecolor='black', alpha=0.7, color='skyblue', density=True)
    axes[i].axvline(pop_mean, color='red', linestyle='dashed', linewidth=2, label=f'Pop Mean: {pop_mean:.2f}')
    axes[i].set_title(f'Sample Size n = {n}')
    axes[i].set_xlabel('Sample Mean')
    axes[i].set_ylabel('Density')
    axes[i].legend()

    simulated_se = np.std(sample_means)
    theoretical_se = pop_std / np.sqrt(n)

    print(f"n = {n} uchun:")
    print(f"  Simulyatsiya qilingan Standard Error (SE): {simulated_se:.4f}")
    print(f"  Nazariy Standard Error (\u03c3/\u221an):        {theoretical_se:.4f}\n")

plt.tight_layout()
plt.show()
```
