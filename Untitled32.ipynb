```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

n, p = 20, 0.3

# 1. Qo'lda hisoblash
E_X_manual = n * p
Var_X_manual = n * p * (1 - p)
sigma_manual = np.sqrt(Var_X_manual)

# 2. Scipy stats orqali tekshirish
dist = stats.binom(n, p)
E_X_scipy = dist.mean()
Var_X_scipy = dist.var()
sigma_scipy = dist.std()

print("--- Qo'lda hisoblangan qiymatlar ---")
print(f"E[X] (Kutish qiymati): {E_X_manual}")
print(f"Var(X) (Dispersiya): {Var_X_manual:.2f}")
print(f"Sigma (O'rtacha kvadratik chetlashish): {sigma_manual:.4f}")

print("\n--- Scipy orqali olingan qiymatlar ---")
print(f"dist.mean(): {E_X_scipy}")
print(f"dist.var(): {Var_X_scipy:.2f}")
print(f"dist.std(): {sigma_scipy:.4f}")

# 3. PMF grafigini chizish
x = np.arange(0, n + 1)
pmf = dist.pmf(x)

plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='skyblue', alpha=0.7, label='Binomial PMF')
plt.axvline(E_X_scipy, color='red', linestyle='--', linewidth=2, label=f'mu ({E_X_scipy})')
plt.axvline(E_X_scipy - sigma_scipy, color='green', linestyle=':', linewidth=2, label=f'mu - sigma ({E_X_scipy - sigma_scipy:.2f})')
plt.axvline(E_X_scipy + sigma_scipy, color='green', linestyle=':', linewidth=2, label=f'mu + sigma ({E_X_scipy + sigma_scipy:.2f})')

plt.title('Binomial Taqsimot B(20, 0.3) PMF va Chegaralar')
plt.xlabel('Muvaffaqiyatlar soni (X)')
plt.ylabel('Ehtimollik')
plt.xticks(x)
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import stats

n_binom, p_binom = 10, 0.4
a, b = 3, 7

# 1-usul: Formuladan hisoblash
var_X_manual = n_binom * p_binom * (1 - p_binom)
var_Y_formula = (a**2) * var_X_manual

# 2-usul: Simulyatsiya orqali hisoblash
np.random.seed(42)
X_sim = np.random.binomial(n_binom, p_binom, size=100_000)
Y_sim = a * X_sim + b
var_Y_simul = np.var(Y_sim)

print("--- Dispersiya Xossalari Isboti ---")
print(f"1-usul (Formuladan): Var(Y) = {var_Y_formula:.4f}")
print(f"2-usul (Simulyatsiyadan): Var(Y) = {var_Y_simul:.4f}")
print(f"Farq: {abs(var_Y_formula - var_Y_simul):.4f}")

# Konstanta b sonini qo'shish barcha nuqtalarni grafikda shunchaki o'ngga suradi, lekin ular orasidagi masofani (sochilishni) o'zgartirmaydi.
# Matematik tildan tushuntirganda: Var(aX + b) = E[((aX + b) - E[aX + b])^2] = E[(aX + b - aE[X] - b)^2] = E[(aX - aE[X])^2] = a^2 * Var(X).
# Ko'rib turganingizdek, 'b' va '-b' qisqarib ketadi, shu sababli b konstanta dispersiyaga mutlaqo ta'sir qilmaydi.
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = [1, 2, 3, 4, 5]
probs = [0.10, 0.20, 0.30, 0.25, 0.20]

# 1. Yig'indini tekshirish
initial_sum = sum(probs)
print(f"Boshlang'ich ehtimolliklar yig'indisi: {initial_sum:.2f}")

# 2. P(X=5) ni tuzatish (chunki yig'indi 1.05 chiqdi, 0.05 ni olib tashlaymiz)
if initial_sum != 1.0:
    probs[4] = 1.0 - sum(probs[:4])

print(f"Tuzatilgandan keyingi ehtimolliklar yig'indisi: {sum(probs):.2f}")
print(f"Yangi PMF ro'yxati: {probs}")

# 3. E[X] ni hisoblash ( x * P(X=x) ning yig'indisi )
E_X = sum(x * p for x, p in zip(x_vals, probs))
print(f"Tuzatilgan PMF bo'yicha E[X] (Kutish qiymati): {E_X:.2f}")

# 4. Grafik chizish
plt.figure(figsize=(8, 5))
plt.bar(x_vals, probs, color='teal', alpha=0.7, edgecolor='black')
plt.axvline(E_X, color='red', linestyle='--', linewidth=2, label=f'E[X] = {E_X:.2f}')

plt.title("To'g'rilangan PMF Ehtimolliklar Taqsimoti")
plt.xlabel('Zar qiymatlari (X)')
plt.ylabel('Ehtimollik (P)')
plt.xticks(x_vals)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

lam = 5
dist = stats.poisson(lam)

# 1. E[X] va Var(X)
E_X = lam
Var_X = lam

# 2. P(X = 0)
p_0 = dist.pmf(0)

# 3. P(X >= 8) -> 1 - P(X <= 7)
p_ge_8 = 1 - dist.cdf(7)  # yoki dist.sf(7) yordamida

print("--- Poisson Taqsimoti Natijalari ---")
print(f"E[X] (Kutish qiymati): {E_X}")
print(f"Var(X) (Dispersiya): {Var_X}")
print(f"P(X = 0) (Hech kim kelmaslik ehtimoli): {p_0:.4f} ({p_0*100:.2f}%)")
print(f"P(X >= 8) (8 va undan ko'p mijoz kelish ehtimoli): {p_ge_8:.4f} ({p_ge_8*100:.2f}%)")

# 4. Grafik chizish
x = np.arange(0, 16)
pmf_vals = dist.pmf(x)

plt.figure(figsize=(10, 6))
plt.stem(x, pmf_vals, linefmt='g-', markerfmt='go', basefmt='r-', label='Poisson PMF')
plt.axvline(lam, color='blue', linestyle='--', linewidth=2, label=f'E[X] = lambda ({lam})')

plt.title('Poisson(lambda=5) Taqsimoti PMF Grafigi')
plt.xlabel('Kelgan mijozlar soni (X)')
plt.ylabel('Ehtimollik')
plt.xticks(x)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy import stats

n, p = 15, 0.4
np.random.seed(42)
X_sim = np.random.binomial(n, p, size=100_000)

# O'rtacha hisoblashlar (Formuladan)
e_x_f = n * p
e_4x_f = 4 * e_x_f
e_4x_10_f = 4 * e_x_f + 10
e_x2_f = (n * p * (1 - p)) + (e_x_f ** 2)  # Var(X) + (E[X])^2

# O'rtacha hisoblashlar (Simulyatsiyadan)
e_x_s = X_sim.mean()
e_4x_s = (4 * X_sim).mean()
e_4x_10_s = (4 * X_sim + 10).mean()
e_x2_s = (X_sim ** 2).mean()

# Natijalarni chiroyli jadval qilib chiqarish
data = {
    "Ifoda": ["E[X]", "E[4X]", "E[4X + 10]", "E[X^2]"],
    "Formuladan (Qo'lda)": [e_x_f, e_4x_f, e_4x_10_f, e_x2_f],
    "Simulyatsiyadan": [e_x_s, e_4x_s, e_4x_10_s, e_x2_s]
}

df_res = pd.DataFrame(data)
print(df_res.to_string(index=False))

# E[X^2] va (E[X])^2 mutlaqo boshqa-boshqa narsalar.
# (E[X])^2 bu — o'rtacha qiymatning shunchaki kvadrati (ya'ni 6^2 = 36).
# E[X^2] esa — har bir qiymatni kvadratga oshirib chiqib keyin o'rtachasini hisoblashdir (bizda u 39.6 chiqdi).
# Ularning orasidagi farq aynan Dispersiyani beradi: Var(X) = E[X^2] - (E[X])^2.
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([48, 50, 52, 49, 51])
B = np.array([10, 20, 90, 80, 50])

print("--- O'yinchi A ---")
print(f"O'rtacha daromad (Mean): {A.mean():.2f}k")
print(f"Dispersiya (Var): {A.var():.2f}")
print(f"Standart chetlashish (Std): {A.std():.2f}")

print("\n--- O'yinchi B ---")
print(f"O'rtacha daromad (Mean): {B.mean():.2f}k")
print(f"Dispersiya (Var): {B.var():.2f}")
print(f"Standart chetlashish (Std): {B.std():.2f}")

# Subplot (1x2) bilan grafik chizish
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

days = ['Kun 1', 'Kun 2', 'Kun 3', 'Kun 4', 'Kun 5']

ax1.plot(days, A, marker='o', color='green', linewidth=2)
ax1.set_title("O'yinchi A ning kunlik daromadi")
ax1.set_ylabel('Daromad (ming so\'m)')
ax1.set_ylim(0, 100)
ax1.grid(True)

ax2.plot(days, B, marker='s', color='orange', linewidth=2)
ax2.set_title("O'yinchi B ning kunlik daromadi")
ax2.set_ylim(0, 100)
ax2.grid(True)

plt.suptitle("O'yinchilar daromadlarining barqarorlik tahlili")
plt.show()

# Har ikki o'yinchining ham o'rtacha daromadi (Mean) bir xil — 50 ming so'm.
# Lekin O'yinchi A ning dispersiyasi (1.6) juda past, ya'ni uning natijalari har kuni bir xil barqaror atrofida bo'ladi.
# O'yinchi B ning dispersiyasi (1040.0) esa dahshatli darajada katta — u juda beqaror (goh 10k, goh 90k topadi).
# Shuning uchun men menejer sifatida O'yinchi A ni tanlayman, chunki biznesda risk past va barqarorlik yuqori bo'lgan variant afzal ko'riladi.
```

```python
import numpy as np
from scipy import stats

mu, sigma = 170, 10
dist = stats.norm(mu, sigma)

# Ehtimolliklarni CDF orqali hisoblash
p1 = dist.cdf(160)
p2 = dist.cdf(180) - dist.cdf(160)
p3 = dist.sf(185)  # Yoki 1 - dist.cdf(185)

# Bonus: 68-95-99.7 qoidasi (1 sigma oralig'i)
rule_68 = dist.cdf(mu + sigma) - dist.cdf(mu - sigma)

print("--- CDF Ehtimollik Natijalari ---")
print(f"P(X <= 160) (160 sm dan past): {p1:.4f} ({p1*100:.2f}%)")
print(f"P(160 < X <= 180) (160 va 180 sm orasi): {p2:.4f} ({p2*100:.2f}%)")
print(f"P(X > 185) (185 sm dan baland): {p3:.4f} ({p3*100:.2f}%)")

print(f"\nUchala ehtimollik yig'indisi: {p1 + p2 + p3:.4f}")
print("Tushuntirish: Bu ehtimolliklarning yig'indisi 1 ga teng chiqmaydi, chunki ular to'liq fazoni qoplamaydi.")
print("Ular bir-biri bilan kesishadigan yoki orasida bo'shliq qoladigan turli xil oraliqlardir (masalan, 180 va 185 oralig'i hisobga olinmay qolib ketgan).")

print(f"\n--- Bonus: 68-95-99.7 Qoidasini Tekshirish ---")
print(f"P(mu-sigma < X < mu+sigma) qiymati: {rule_68:.4f} ({rule_68*100:.2f}%)")
```
