```python
 import numpy as np
import matplotlib.pyplot as np_plot # Grafik uchun

# 1. Boshlang'ich holat
w = 10.0  # Modelning tasodifiy og'irligi (tog'ning juda baland joyida turibdi)
learning_rate = 0.1  # Limitdan kelgan kichik qadam kattaligi

# Loss funksiyasi: L(w) = w^2 (Eng kam xato w = 0 bo'lganda bo'ladi)
history_w = [w]

print("--- O'QITISH BOSHLANDI ---")
for epoch in range(10):
    # Loss (Xatolik)
    loss = w ** 2

    # HOSILA (Gradient): d(w^2)/dw = 2*w
    # Bu bizga qiyalik va yo'nalishni beradi!
    derivative = 2 * w

    # Og'irlikni to'g'rilash (Gradient Descent adashtirmasdan tushiradi)
    w = w - learning_rate * derivative
    history_w.append(w)

    print(f"Qadam {epoch+1}: Loss = {loss:.4f}, Hosila (Qiyalik) = {derivative:.4f}, Yangi w = {w:.4f}")

print("\nModel minimal xatoga erishdi! Yakuniy w ~ 0 ga teng bo'ldi.")
```

```python
import sympy as sp

x = sp.Symbol('x')

# b)
f_b = 7*x**3 + x - 100
print("b) f'(x) =", sp.diff(f_b, x))

# c)
f_c = (x**5)/5 - x
df_c = sp.diff(f_c, x)
print("c) f'(x) =", df_c)
print("c) f'(x) = 0 ildizlari:", sp.solve(df_c, x))

# d)
f_d = 1/(x**2)
print("d) f'(x) =", sp.diff(f_d, x))
```

```python
# Loss funksiyamiz 2 ta og'irlikka bog'liq: Loss = w1^2 + 3*w2^2

def compute_loss(w1, w2):
    return w1**2 + 3 * (w2**2)

# Boshlang'ich og'irliklar
w1 = 4.0
w2 = 3.0

# 1. Xususiy hosilalarni matematik hisoblash:
# dLoss/dw1 = 2 * w1
# dLoss/dw2 = 6 * w2
d_w1 = 2 * w1  # 2 * 4 = 8
d_w2 = 6 * w2  # 6 * 3 = 18

print(f"Hozirgi Loss: {compute_loss(w1, w2)}")
print(f"w1 bo'yicha xususiy hosila (dLoss/dw1): {d_w1}")
print(f"w2 bo'yicha xususiy hosila (dLoss/dw2): {d_w2}")

# Gradient bo'yicha qadam tashlaymiz (Learning rate = 0.1)
lr = 0.1
w1_new = w1 - lr * d_w1  # 4 - 0.8 = 3.2
w2_new = w2 - lr * d_w2  # 3 - 1.8 = 1.2

print(f"\nYangi w1: {w1_new}, Yangi w2: {w2_new}")
print(f"Yangi kamaytirilgan Loss: {compute_loss(w1_new, w2_new):.2f}")
```

```python
import numpy as np


def gradient(f, x, y, h=1e-5):
    gx = (f(x + h, y) - f(x, y)) / h  # x bo'yicha
    gy = (f(x, y + h) - f(x, y)) / h  # y bo'yicha
    return np.array([gx, gy])


def f(x, y):
    return x**2 + y**2


print(gradient(f, 1, 2))  # [2. 4.]
print(gradient(f, 0, 0))  # [0. 0.]

# (0,0) da gradient NOL vektor -
# «kosa» tubidamiz, hamma tomon tekis!
```

```python
# Loss funksiyamiz: Loss = w^2
# Hosilasi (Gradient): dLoss/dw = 2*w

w = 10.0          # Boshlang'ich og'irlik (tog'ning tepasidamiz)
learning_rate = 0.1 # Qadam kattaligi
epochs = 10       # Necha marta qadam tashlashimiz

print("--- GRADIENT DESCENT BOSHLANDI ---")

for epoch in range(1, epochs + 1):
    # 1. Loss ni hisoblaymiz
    loss = w ** 2

    # 2. Gradient (Hosila/Qiyalik) ni hisoblaymiz
    gradient = 2 * w

    # 3. GRADIENT DESCENT FORMULASI bo'yicha og'irlikni yangilaymiz:
    w = w - learning_rate * gradient

    print(f"Qadam {epoch}: Loss = {loss:.4f}, Gradient = {gradient:.4f}, Yangi w = {w:.4f}")

print("\nNatija: Model eng tubiga yetib keldi! w deyarli 0 ga teng bo'ldi.")
```
