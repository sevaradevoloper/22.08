```python
#   DARAJA 4  Matritsa ko'paytmasi:  * ≠ @   ★★★   Mustahkamlash
# Maqsad: element-wise ko'paytma va haqiqiy matritsa ko'paytmasini aralashtirmaslik.
# 4.1  Farqni ko'ring
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])
print(A)
print(B)
# A * B va A @ B ni alohida chop eting. Ikkalasi ham 2×2, lekin natijalar boshqacha. Farqini o'z so'zingiz bilan izohlang.
# 4.2  Shape moslik — oldindan ayting
# Quyidagilardan qaysilari ISHLAYDI, qaysilari XATO beradi? Ishlaydiganlari uchun natija shape'ini yozing. Avval javob yozing, keyin kod bilan tekshiring:
# (2 x 3) @ (3 x 4)
# (3 x 2) @ (3 x 4)
# (4 x 2) @ (2 x 5)
# (2 x 2) @ (2 x 2)

# →  Eslatma: ichki o'lchamlar teng bo'lishi shart — (m × k) @ (k × n).


# 4.3  Transpose bilan tuzatish
# A = [[1,2,3],[4,5,6]] (shape 2×3). A @ A ishlamaydi — nega? A @ A.T va A.T @ A ni sinab ko'ring. Har birining natija shape'i qanday?


print('N  4.1 -------------------------------------------')
print(A * B) #bu yerda togridan togri kopaytiradi A ni [1, 2] si B [2, 0] ga 1* 2 ,2* 0 ga
print(A @ B) #a @ esa A qator * B ustub 1*2 2*1 = 4 , 3*2 4* 1 = 10 shunda yangi birinchi qator [4 10 ]chiqdi
print('N  4.2 -------------------------------------------')
# (2 x 3) @ (3 x 4) ishlidi k = 3
# (3 x 2) @ (3 x 4) ishlamidi k teng emas
# (4 x 2) @ (2 x 5) ishlidi k = 2 teng
# (2 x 2) @ (2 x 2)ishlidi k = 2 teng

# 1. (2 x 3) @ (3 x 4) -> ISHlIDI
A1 = np.zeros((2, 3))
B1 = np.zeros((3, 4))
print("1) Shape:", (A1 @ B1).shape)

# 2. (3 x 2) @ (3 x 4) -> XATO BERADI
# A2 = np.zeros((3, 2))
# B2 = np.zeros((3, 4))
# print((A2 @ B2).shape)

# 3. (4 x 2) @ (2 x 5) -> ISHlIDI
A3 = np.zeros((4, 2))
B3 = np.zeros((2, 5))
print("3) Shape:", (A3 @ B3).shape)

# 4. (2 x 2) @ (2 x 2) -> ISHlIDI
A4 = np.zeros((2, 2))
B4 = np.zeros((2, 2))
print("4) Shape:", (A4 @ B4).shape)

print('N  4.3 -------------------------------------------')
# Transpose bilan tuzatish bu matritsalani ustida amal bajarvotan paytimiz agar nobodo shape togri keme qolsa qatori ustunga ,ustuni qatorga ozgartirb beruvchi func
A = np.array([[1,2,3],[4,5,6]])
print('A before shape: ' , A.shape)
A_new = A.T
print('A after shape: ' , A_new.shape)
```

```python
#   DARAJA 5  ML tatbiqi   ★★★   Mustahkamlash
# Maqsad: nazariyani haqiqiy misolda ishlatish.

# 5.1  Linear regression
X = np.array([[25, 3.2],
              [34, 5.1],
              [41, 6.8]])   # yosh, daromad
w = np.array([1.5, 20])
b = 10

# y = X @ w + b ni hisoblang. Natijada nechta son bor? Nega aynan shuncha? Birinchi obyekt uchun natijani qo'lda tekshiring.
# 5.2  Broadcasting
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# (a) Har bir elementga 100 qo'shing. (b) v = [10, 20, 30] vektorini A ning har bir qatoriga qo'shing (A + v). Ikkala natijani ham oldindan ayting, keyin tekshiring.

# 5.3  Kosinus o'xshashligi
#  Kosinus o'xshashligi (Cosine Similarity) — bu ikkita vektor bir-biriga geometrik jihatdan qanchalik mos kelishini (bir tomonga yo'nalganini) o'lchaydigan ko'rsatkich.
# U Machine Learning'da (ayniqsa Tavsiya tizimlari / Recommendation Systems va NLP - matnlar bilan ishlashda) juda ko'p ishlatiladi.

# Uchta mijozning xarid vektori berilgan:
m1 = np.array([5, 0, 3, 1])
m2 = np.array([4, 0, 2, 1])
m3 = np.array([0, 5, 1, 4])

# Formula:  cos_sim(a, b) = (a @ b) / (||a|| * ||b||)
# m1–m2 va m1–m3 o'xshashligini hisoblang. Qaysi ikki mijoz bir-biriga ko'proq o'xshash? Nega, deb o'ylaysiz?


print('N  5.1-------------------------------------------')
# X @ w = 452 +10 = 462 chiqdi bu umumiy,agar 1 chi odamiki : y1 = (25⋅1.5)+(3.2⋅20)=37.5+64=101.5  chiqadi
print('N  5.2-------------------------------------------')
# broadcasting. bu bizaga matritsani chozib beradi Mat + vektor qib yoki skalyar
a = A + 100
print(a)
v = np.array([10, 20, 30])
b= A+v
print(a)
print(b)
print('N  5.3-------------------------------------------')
# Normlar (uzunliklar)
norm1 = np.linalg.norm(m1) # sqrt(25 + 0 + 9 + 1) = sqrt(35) ≈ 5.916
norm2 = np.linalg.norm(m2) # sqrt(16 + 0 + 4 + 1) = sqrt(21) ≈ 4.582
norm3 = np.linalg.norm(m3) # sqrt(0 + 25 + 1 + 16) = sqrt(42) ≈ 6.480

# m1 va m2 o'xshashligi
sim_1_2 = (m1 @ m2) / (norm1 * norm2)
# m1 @ m2 = (5*4 + 0*0 + 3*2 + 1*1) = 27
# sim_1_2 = 27 / (5.916 * 4.582) ≈ 0.996

# m1 va m3 o'xshashligi
sim_1_3 = (m1 @ m3) / (norm1 * norm3)
# m1 @ m3 = (5*0 + 0*5 + 3*1 + 1*4) = 7
# sim_1_3 = 7 / (5.916 * 6.480) = 0.182

print(f"m1 va m2 o'xshashligi: {sim_1_2:.3f}")
print(f"m1 va m3 o'xshashligi: {sim_1_3:.3f}")
# m1 va m 2 orasidagi o'xshashlik: 0.996 (Deyarli 1 ga teng, ya'ni juda yaqin!)
# m1 va m 3 orasidagi o'xshashlik: 0.182 (Nolga yaqin, ya'ni o'xshamaydi)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Matritsani aniqlaymiz
N = np.array([[2, 4],
              [1, 2]])

# Eigenvalues va Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(N)
v1 = eigenvectors[:, 0]  # lambda = 0 ga mos eigenvector
v2 = eigenvectors[:, 1]  # lambda = 4 ga mos eigenvector

# Test uchun panjara (grid) nuqtalarini yaratamiz
x = np.linspace(-2, 2, 9)
y = np.linspace(-2, 2, 9)
X, Y = np.meshgrid(x, y)
points = np.vstack([X.ravel(), Y.ravel()])

# Matritsa ta'siridan keyingi yangi nuqtalar (N @ points)
transformed_points = N @ points

# Grafik chizish
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --- 1-GRAFIK: Transformatisyadan OLDIN ---
ax1 = axes[0]
ax1.scatter(points[0, :], points[1, :], color='royalblue', alpha=0.5, label='2D Fazo nuqtalari')
# Eigenvector yo'nalishlari
ax1.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='gold', linewidth=2, label=f'v1 (λ={eigenvalues[0]:.0f}) yo\'nalishi')
ax1.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='limegreen', linewidth=2, label=f'v2 (λ={eigenvalues[1]:.0f}) yo\'nalishi')

ax1.set_xlim(-10, 10)
ax1.set_ylim(-10, 10)
ax1.axhline(0, color='grey', linestyle='--', alpha=0.5)
ax1.axvline(0, color='grey', linestyle='--', alpha=0.5)
ax1.grid(True, linestyle=':', alpha=0.6)
ax1.set_title("1. Dastlabki 2D Fazo (Dastlabki nuqtalar va vektorlar)", fontsize=12)
ax1.legend(loc='upper left')

# --- 2-GRAFIK: Transformatisyadan KEYIN (N @ v) ---
ax2 = axes[1]
ax2.scatter(transformed_points[0, :], transformed_points[1, :], color='crimson', alpha=0.7, label='1D Chiziqqa tushgan nuqtalar')

# O'zgargan Eigenvectorlar
v1_transformed = N @ v1
v2_transformed = N @ v2

ax2.quiver(0, 0, v1_transformed[0], v1_transformed[1], angles='xy', scale_units='xy', scale=1, color='gold', linewidth=2, label='N @ v1 = 0 (Yo\'q bo\'ldi)')
ax2.quiver(0, 0, v2_transformed[0], v2_transformed[1], angles='xy', scale_units='xy', scale=1, color='limegreen', linewidth=2, label='N @ v2 = 4 * v2 (4x cho\'zildi)')

ax2.set_xlim(-10, 10)
ax2.set_ylim(-10, 10)
ax2.axhline(0, color='grey', linestyle='--', alpha=0.5)
ax2.axvline(0, color='grey', linestyle='--', alpha=0.5)
ax2.grid(True, linestyle=':', alpha=0.6)
ax2.set_title("2. N Matritsasi ta'siridan KEYIN (Fazo 1D chiziqqa g'ijimlandi)", fontsize=12)
ax2.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# 2. Qo'lda hisoblang → NumPy da tekshiring   [qo'lda + kod]
# Avval daftarda hisoblang, keyin kodda tasdiqlang. Natija mos kelmasa — nega, deb izohlang.
# •  [[3, 0], [0, 2]] @ [2, -1] (3⋅2)+(0⋅−1)=6+0=6 (0⋅2)+(2⋅−1)=0−2=−2 [6 -2]
# •  [[0, -1], [1, 0]] @ [2, 3] (0⋅2)+(−1⋅3)=0−3=−3 (1⋅2)+(0⋅3)=2+0=2 [-3 2 ]
# •  [[1, 0], [0, -1]] @ [-4, 5] (1⋅−4)+(0⋅5)=−4+0=−4 (0⋅−4)+(−1⋅5)=0−5=−5 [-4 -5 ]
# •  [[1, 2], [0, 1]] @ [1, 3]. (1⋅1)+(2⋅3)=1+6=7 (0⋅1)+(1⋅3)=0+3=3 [7 3]
# 1-misol
res1 = np.array([[3, 0], [0, 2]]) @ np.array([2, -1])
print("1-natija:", res1)

# 2-misol
res2 = np.array([[0, -1], [1, 0]]) @ np.array([2, 3])
print("2-natija:", res2)
# 3-misol
res3 = np.array([[1, 0], [0, -1]]) @ np.array([-4, 5])
print("3-natija:", res3)

# 4-misol
res4 = np.array([[1, 2], [0, 1]]) @ np.array([1, 3])
print("4-natija:", res4)
```
