````python
import numpy as np

# 1. Boshlang'ich qadamlarimiz (Vektorlar)
# [X o'qi (Sharq), Y o'qi (Shimol)]
v1 = np.array([1, 0])  # Sharqqa 1 qadam (X=1, Y=0)
v2 = np.array([0, 1])  # Shimolga 1 qadam (X=0, Y=1)

print(f"v1 (Sharq): {v1}")
print(f"v2 (Shimol): {v2}\n")

# 2. Bizga berilgan topshiriq (Koeffitsiyentlar): 2v1 + 3v2
# Sharqqa 2 marta, Shimolga 3 marta ko'paytir degani
c1 = 2
c2 = 3

# 3. Chiziqli kombinatsiyani hisoblaymiz (Vektorlarni ko'paytirib, qo'shamiz)
yangi_nuqta = (c1 * v1) + (c2 * v2)

# Ichki hisob-kitob bunday bo'ladi:
# (2 * [1, 0]) + (3 * [0, 1]) = [2, 0] + [0, 3] = [2, 3]

print("--- CHIZIQLI KOMBINATSIYA NATIJASI ---")
print(f"Formula: {c1}*v1 + {c2}*v2")
print(f"Siz yetib borgan yangi nuqta (koordinata): {yangi_nuqta}")
print(f"Ma'nosi: Sharqqa {yangi_nuqta[0]} qadam va Shimolga {yangi_nuqta[1]} qadam yurdingiz!")
````
