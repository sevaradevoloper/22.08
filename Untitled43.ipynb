```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
sns.set_theme(style='whitegrid') #grafik korinish ni chiroyli qilish uchn
```

```python
# 1 Bosqich
#datasetni yuklab olamiz

df = pd.read_csv('/content/uzum_orders (1).csv')
print(df.isnull().sum()) #null qiymatlari bor yoqliginii aniqlimiz
print(df.shape) #nechda qator va ustun borligi
print(df.info())#umumiy malumotlarnii qaytaradi
df
```

```python

# 2 Bosqich
#nulllarni toldirish

df['reyting'] = df['reyting'].fillna(df['reyting'].median()) #reyting ustunida null qiymatlarni kategoriyasiga qarab median bilan toldirb chiqtik
print(df.isnull().sum()) #null qiymatlari endi ham bormi yoqmi tekshirib olamiz
print(df.head())    #boshdagi 5 ta qatorni olam
```

```python
#3 Bosqich Outlierlarni aniqlar(narx)
#EDA (Exploratory Data Analysis) — bu ma'lumotlarni quruq tahlil qilishdan oldin, ularni "shunchaki varoqlab, razvedka qilib chiqish" bosqichi.

df.describe() #describe orqali uni umumiy malumotlarini korb chiqamz
#mean va median ga qarab qaysi ustunda outlierlar bor
#narx ustunida outlierlar bor ekan sababi mean(143.67) > median(132.00)
#narx ustunini endi outlierlarini tozalaymiz

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

#boxplotda koramz

sns.boxplot(data=df,y='narx',ax=axes[0],color='pink')
axes[0].set_title("Tozalashdan oldingi Narxlar (Outlierlar bor)")
axes[0].set_ylabel("Narx (ming so'm)")



#4 Bosqich Outlierlarni tozalash


Q1 = df['narx'].quantile(0.25)
Q3 = df['narx'].quantile(0.75)
print(Q1)
print(Q3)
IQR = Q3- Q1
print(IQR)
pastki_chegara = Q1 - 1.5 * IQR
yuqori_chegara = Q3 + 1.5 * IQR
print(f"pastki_chegara: {pastki_chegara}")
print(f"yuqori_chegara : {yuqori_chegara}")

outliers = df[(df['narx'] < pastki_chegara) | (df['narx'] > yuqori_chegara)]
df = df[(df['narx'] >= pastki_chegara) & (df['narx'] <= yuqori_chegara)]

print(f"Topilgan outlierlar (keskin qiymatlar) soni: {len(outliers)} ta")
print(f"Tozalangandan keyingi qatorlar soni: {len(df)}")
print(f"Yangi Mean (O'rtacha): {df['narx'].mean():.2f}")
print(f"Yangi Median (Mediana): {df['narx'].median():.2f}")
#tozalanganini endi ko'ramiz
plt.figure(figsize = (8,5))
sns.boxplot(data=df,y='narx',ax=axes[1],color ='green')
axes[1].set_title("Tozalashdan keyingi Narxlar (Outlierlar bor)")
axes[1].set_ylabel("Narx (ming so'm)")
plt.show()
```

```python
#HISTPLOT() orqali median vs meani qanchali bir birga yaqnligini korib olamz va outlierla qanchali kamayganini


plt.figure(figsize = (8,5))
mean_narx = df['narx'].mean()
median_narx = df['narx'].median()
sns.histplot(
    data=df,
    x='narx',
    kde=True, #line chiziq
    bins=30 #30ta bolak
)
plt.axvline(x=mean_narx, color='red', linestyle='--', linewidth=2, label=f"Mean: {mean_narx:.1f}")

# 2. Median (Mediana) uchun yashil chiziq qo'shish
plt.axvline(x=median_narx, color='green', linestyle='-', linewidth=2, label=f"Median: {median_narx:.1f}")

# Grafikga tushuntirish (Legend) qo'shish, chiziqlar nimaligini ko'rsatish uchun
plt.legend()
plt.title("Narxlar taqsimoti va ularning Mean/Median chiziqlari")
plt.show()


#Xulosa:MEAN vs MEDIDAN oraligi yaqn ya'ni outlierlar kamaydi
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

#scatterplotda
sns.scatterplot(
    data=df,
    y='narx',
    x='mijoz_yoshi',
    alpha=0.4,     # Nuqtalarni shaffof qilish (ustma-ust tushganini ko'rsatadi)
    color='purple',
    ax=axes[0]
)
axes[0].set_title("Mijoz yoshiga qarab uylar narxi")

#Xulosa corr pachti yoq yani naxrga mijoz yoshi tasir qilmaydi


#boxplotda

sns.boxplot(
    data=df,
    x='filial',
    y='narx',
    order=df['filial'].value_counts().index, palette='viridis',
    ax=axes[1]
)

axes[1].set_title("Filiyallar bo'yica uylar narxi",fontsize=17)
#Xulosa : Filallarda deyarli outlierlar yoq ,median pachti bir xil :OKAY

# X o'kidagi yozuvlarni (agar filiallar bo'lsa) 20 gradusga qiyshaytirishning TO'G'RI YO'LI:
axes[1].tick_params(axis='x', rotation=20)
plt.tight_layout()
plt.show()
```

```python
#Endi corr() solishtrb koramz va heatmap orqali koramz

fig, axes = plt.subplots(1, 2, figsize=(16, 6))


sns.scatterplot(
    data=df,
    x='yetkazish_kuni',
    y='reyting',
    alpha=0.4,
    color='purple',
    ax=axes[0]
)

r = df['yetkazish_kuni'].corr(df['reyting'])
print(f"r = {r:.2f}")


#heatmap

sonli = df[['narx','yetkazish_kuni','mijoz_yoshi','reyting',]]

corr=sonli.corr()
sns.heatmap(
    corr,
    annot=True,
    cmap='RdYlGn',
    center=0,
    fmt='.2f',
    square=True,
    ax=axes[1]
)
axes[1].set_title('korrelyatsiya matritsasi')
plt.tight_layout()
plt.show()

#Hulosa: Yetkazish va Reyting : Spearman corr ga kirvoti sababi bittasi osa ikkinchisi tushadi
# qolgan ustunlarda esa bogliklik detarli yo'q
```

```python
# Shapiro-Wilk normallik testi
stat, p = stats.shapiro(
    df['narx'].sample(500, random_state=42))

print(f'p-value = {p:.4f}')
# p-value = 0.0000




# Oʻsha "95% va 5%" qoidasi nima?
# Biz boshida shunday shart qoʻyamiz:
# "Biz Uzum ma'lumotlariga 95% ishonch bilan qaraymiz. Agar ma'lumotlarimizda gʻalati, kutilmagan,
# normal holatdan butunlay farq qiladigan holat uchrash ehtimoli 5% dan kam boʻlsa, demak bu ma'lumotda aniq bir "ayb" (qiyshiqlik, outlier) bor deb hisoblaymiz."
# Mana shu 5% — statistikada α=0.05 deb yoziladi.


if p < 0.05:
    print('Normal EMAS')
else:
    print('Normalga yaqin')

# p-value nima o'zi? (Nega 0.0000 chiqsa Normal emas?)
# p-value — bu "Tasodif ehtimolligi". Ya'ni, "Ushbu ma'lumotlarning bunday qiyshiq bo'lib qolishi shunchaki tasodifmi?" degan savolga javob.
# Sizda p-value = 0.0000 (ya'ni 0%) chiqdi.
# Bu degani: Ma'lumotlarning ideal (Normal) holatdan ogʻib, buzilib ketishi tasodif emas! Buning ehtimoli 0%! Demak ma'lumotlar haqiqatdan ham buzilgan, qiyshiq va ichida outlierlar bor — ya'ni Normal EMAS!



#Agar sizda p-value = 0.60 (60%) chiqqanida edi:
# Hulosa:Uzum do'konidagi mahsulotlarimizning narxi Normal taqsimlanmagan.
# Bunga sabab — narxlar ichida o'ta yuqori bo'lgan g'ayritabiiy qiymatlar (outlierlar) mavjudligi va grafikning o'ngga qarab cho'zilganidir (Right-skewed).


# Bu grafik (Normal holatga) umuman o'xshamaydi! Uning ideal qo'ng'iroq bo'lish ehtimoli 0% (p-value = 0.0000)

#agar kamida 5% dan oshasa normal taqsimotga otqazamiz agar shu 5% dan kam bosa u normalga uje kirmidi
```

```python
#Confidence Interval — CI) 95% Ishonch oraligʻi

# Siz Uzum market egasisiz. Sizda jami 1 millionta xaridor bor. Lekin qoʻlingizdagi datasetda bor-yoʻgʻi 1 000 ta xaridorning maʼlumoti bor (bu maʼlumotlar toʻplami namuna yoki sample deyiladi).
# Siz 1 000 ta odamning maʼlumotiga qarab, Uzum marketdagi hamma (1 millionta) odam oʻrtacha necha pul sarflayotganini bilmoqchisiz. Mana shu kod aynan shu ishni qiladi.

narx = df['narx']
n = len(narx)
mean = narx.mean() #Qoʻlingizdagi 1 000 ta odamning oʻrtacha xarid narxi. Masalan, u 135.5 ming soʻm chiqdi.
se = stats.sem(narx)  # Standart xato Bu sizning "adashish koeffitsiyentingiz".
#Ya'ni, 1 000 ta odamdan olingan oʻrtacha qiymat (135.5) butun boshli 1 millionta odamning haqiqiy oʻrtacha narxidan qanchalik uzoqqa adashishi mumkinligini hisoblab beradigan koʻrsatkich.

# 95% ishonch oralig'i (ci)
# loc=mean bo'lishi kerak, minus (-) emas!
ci = stats.t.interval(0.95, df=n-1, loc=mean, scale=se)  #Menga shunday bir chegara (diapazon) topib berki,
#men 95% ishonch bilan butun Uzum marketning haqiqiy o'rtacha narxi shu oraliqqa tushadi deb ayta olay
#(132.9, 138.0). qatytaradi



# f-string bn
print(f'O\'rtacha narx: {mean:.1f}')
print(f'95% Ishonch oralig\'i (CI): {ci}')



#Xulosa:
# Hisob-kitoblarimiz shuni ko'rsatdiki, qo'limizdagi namunaning o'rtacha narxi 135.5 ming so'm.
# Biz 95% ishonch bilan ayta olamizki, Uzum marketdagi barcha xaridorlarning (general majmuaning) haqiqiy o'rtacha narxi 132.9 va 138.0 ming so'm oralig'ida joylashgan.
```

```python
# Chilonzor va Yunusobod filiallari narxlarini ajratib olamiz
# (Ustun nomidagi bo'shliqlar to'g'rilandi va tenglik belgisi '==' qilindi)
chil = df[df['filial'] == 'Chilonzor']['narx']
yunu = df[df['filial'] == 'Yunusobod']['narx']

# Ikki mustaqil namuna uchun T-test (Independent T-test)
t_stat, p = stats.ttest_ind(chil, yunu)

print(f't-statistika = {t_stat:.2f}')
print(f'p-value = {p:.4f}')

# Endi boyagi o'rgangan Alpha (0.05) qoidamizni ishlatamiz:
alpha = 0.05
if p < alpha:
    print("Xulosa: Filiallar o'rtasidagi farq tasodifiy emas! Narxlar sezilarli darajada farq qiladi.")
else:
    print("Xulosa: Filiallar o'rtasidagi farq shunchaki tasodif. Narxlar deyarli bir xil.")
```
