```python
# 6 chi bosqich. Mini loyiha — Uy narxini bashorat qilish

X_uylar = np.array(
    [[60, 2], [85, 3], [45, 1], [120, 4], [70, 2]]  # [maydoni, xonala]
)
w_uylar = np.array([15, 30])
b_uylar = 50
narxlar = X_uylar @ w_uylar + b_uylar
print(" 5 ta uyning hisoblangan narxlari:", narxlar)


eng_qimmat_idx = np.argmax(narxlar)
print(
    f"Eng qimmat uy indeksi: {eng_qimmat_idx}, narxi: {narxlar[eng_qimmat_idx]} mln"
)

qavat = np.array([[3], [5], [2], [9], [4]])
X_yangi = np.hstack((X_uylar, qavat))  # 5x3 matritsa boldi
w_yangi = np.array([15, 30, 10])  # qavat vazni ham qoshildi
narx_yangi = X_yangi @ w_yangi + b_uylar
print("Qavat qo'shilgandan keyingi yangi narxlar:", narx_yangi)
```
